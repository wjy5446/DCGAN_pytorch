```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torchvision

from model import Generator
from utils.misc import load_model
```

```python
def imshow(img, title, size=(5, 5)):
    npimg = img.numpy()
    fig = plt.figure(figsize=size)
    
    for i in range(npimg.shape[0]):
        plt.subplot(1, npimg.shape[0], i+1)
        plt.imshow(npimg[i])
        plt.title(title)
    plt.show()
```

```python
g = Generator()
weight = load_model('model', 'GAN', 'loss')
g.load_state_dict(weight)
```

```python
g.eval()
noise = torch.randn(10, 100, 1, 1)
output = g(noise)
```

```python
imshow(torchvision.utils.make_grid(output.squeeze().data.cpu(), normalize=True), "", size=(10,10))
```
